```python
# Import necessary libraries
import numpy as np
import pandas as pd
```

```python
#load the dataset
df=pd.read_csv("hotel_bookings.csv")
df.head()
```

```python
df.info()
```

```python
#print all columns from dataset
print(df.columns)
```

```python
#find any null values
df.isnull().sum()
```

```python
#drop the null columns 'agent' and 'company' where it contains more null values
df1=df.drop(["agent","company"],axis=1)
```

```python
df1.isnull().sum()
```

```python
#describe the dataset to know the count,mean,std,min,max
df.describe()
```

```python
#select all category data
category=df1.select_dtypes(include=["object"]).copy()
category.head()
```

```python
print(category.isnull().sum())
```

```python
category.country.unique()
```

```python
category.country.mode()
```

```python
# fill null values in category data using mode
category.fillna(category['country'].mode()[0],inplace=True)
category.isnull().sum()
```

```python
#print unique values of category data of every column
print(category.apply(pd.unique))
```

```python
# Make a copy of the 'category' named 'encode'
encode=category.copy()
encode.head()
```

```python
# Performing encoding based on whether 'hotel' column contains "City Hotel"
encode["hotel"]=np.where(encode["hotel"].str.contains("City Hotel"),1,0)
encode.sample(10)
```

```python
#import labelencoder
from sklearn.preprocessing import LabelEncoder
# Create a LabelEncoder instance
lb=LabelEncoder()
category_country=encode.copy()
```

```python
# Transform the 'country' column into numerical values
category_country["country_code"]=lb.fit_transform(encode["country"])
category_country.sample(10)
```

```python

# Apply get_dummies to the 'arrival_date_month' column
encoded_months = pd.get_dummies(encode['arrival_date_month'], prefix='month')

# Concatenate the encoded months with the original DataFrame
encoded_df = pd.concat([encode, encoded_months], axis=1)

# Drop the original 'arrival_date_month' column if needed
# encoded_df.drop('arrival_date_month', axis=1, inplace=True)

# Display the resulting DataFrame
print(encoded_df)
```
